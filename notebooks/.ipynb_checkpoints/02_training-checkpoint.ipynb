# üìà Baseline Model Training and Evaluation

In this section, we establish a performance baseline for our intrusion detection system. We will train two simple but effective models‚Äî**Logistic Regression** and **Random Forest Classifier**‚Äîto understand how well they perform on the UNSW-NB15 dataset.

We will evaluate their performance using **accuracy** and **F1-score**, with a specific focus on the F1-score as it is a more reliable metric for our imbalanced dataset. We'll use the pre-processed data and the stratified train-test split defined in the previous steps.

```python
# ====================================================================================================
# CELL 4: Baseline Model (Optimized with Sampling Toggle)
# ====================================================================================================

import pandas as pd
from sklearn.metrics import accuracy_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

print("\nTraining baseline models (optimized for imbalanced data)...")
print("=" * 55)

baseline_results = {}
TASK_TYPE = 'binary'  # Assuming a binary classification task
DEBUG_MODE = True     # üîπ Set to False to train on full dataset later

# --- Sampling Logic ---
if DEBUG_MODE:
    sample_size = 0.2  # 20% sample for speed
    print(f"‚ö° DEBUG_MODE ON: Using {sample_size*100:.0f}% of training data")
    X_train_small, _, y_train_small, _ = train_test_split(
        X_train, y_train, train_size=sample_size, random_state=42, stratify=y_train
    )
else:
    print("üöÄ DEBUG_MODE OFF: Using full training data")
    X_train_small, y_train_small = X_train, y_train

# 1. Logistic Regression Baseline
print("\n1. Training Logistic Regression...")
lr_model = LogisticRegression(random_state=42, max_iter=1000, solver='lbfgs')
lr_model.fit(X_train_small, y_train_small)
lr_pred = lr_model.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_pred)
lr_f1 = f1_score(y_test, lr_pred, average='binary' if TASK_TYPE == 'binary' else 'macro')
baseline_results['Logistic Regression'] = {'accuracy': lr_accuracy, 'f1': lr_f1}
print(f"   Accuracy: {lr_accuracy:.4f}, F1-Score: {lr_f1:.4f}")

# 2. Random Forest Baseline (with SMOTE)
print("\n2. Training Random Forest (with SMOTE)...")

# Apply SMOTE to the sampled training data
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_small, y_train_small)

# Train Random Forest
rf_model = RandomForestClassifier(
    n_estimators=30,      # fewer trees for speed
    max_depth=15,         # limit depth
    random_state=42,
    n_jobs=-1
)
rf_model.fit(X_train_resampled, y_train_resampled)
rf_pred = rf_model.predict(X_test)

rf_accuracy = accuracy_score(y_test, rf_pred)
rf_f1 = f1_score(y_test, rf_pred, average='binary' if TASK_TYPE == 'binary' else 'macro')
baseline_results['Random Forest'] = {'accuracy': rf_accuracy, 'f1': rf_f1}
print(f"   Accuracy: {rf_accuracy:.4f}, F1-Score: {rf_f1:.4f}")

# üìä Summary
print(f"\nüìä Baseline Results Summary:")
for model, metrics in baseline_results.items():
    print(f"{model}: Accuracy={metrics['accuracy']:.4f}, F1={metrics['f1']:.4f}")
```

## üìä Baseline Results Summary

The baseline models provide a clear reference point for our final model.

* **Logistic Regression** achieved an **Accuracy of 0.9911** and an **F1-score of 0.9075**. This high F1-score indicates the model is very effective at classifying both normal and attack traffic.
* **Random Forest** achieved an **Accuracy of 0.9756** but a lower **F1-score of 0.7983**. While still a good result, this suggests that the model trained on the SMOTE-resampled data is slightly less effective than the Logistic Regression model, particularly in balancing precision and recall.

Based on these results, our goal is to build a more advanced model that can at least match, if not surpass, the performance of the Logistic Regression baseline.

```python
# ==========================================================================================
# Main Training Workflow
# ==========================================================================================

print("Libraries and custom modules imported successfully!")
print("Original data shape:", X_train.shape, y_train.shape)

# Create a smaller, representative sample of the data (e.g., 10%)
# This prevents your PC from crashing
X_sample, _, y_sample, _ = train_test_split(X_train, y_train, test_size=0.9, random_state=42, stratify=y_train)

print("Sampled data shape:", X_sample.shape, y_sample.shape)

# Use the preprocessor function from your 'preprocess.py' file
preprocessor = create_simple_numeric_preprocessor()

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Choose scoring based on the number of unique classes
n_classes = len(np.unique(y_sample))
scoring = ['accuracy', 'f1'] if n_classes == 2 else ['accuracy', 'f1_weighted']

# Build the final machine learning pipeline
rf_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42, n_jobs=3))
])

# Run cross-validation on the smaller, sampled data
cv_res = cross_validate(rf_pipeline, X_sample, y_sample, cv=skf, scoring=scoring, return_train_score=False, n_jobs=3)

print("\nRF CV results on sampled data:")
print("Accuracy mean:", cv_res['test_accuracy'].mean())
print("F1 mean:", cv_res['test_f1'].mean())
```

```python
# Cell 4: evaluate a set of models on a small sample


models = {
    'RandomForest': RandomForestClassifier(random_state=42, n_jobs=3), # 3 cores
    'HistGB': HistGradientBoostingClassifier(random_state=42)
}
if XGBOOST_AVAILABLE:
    models['XGBoost'] = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42, n_jobs=3) # 3 cores
if LIGHTGBM_AVAILABLE:
    models['LightGBM'] = lgb.LGBMClassifier(random_state=42, n_jobs=3) # 3 cores
models['MLP'] = MLPClassifier(random_state=42, max_iter=30)



results = {}
for name, clf in models.items():
    print("Evaluating", name)
    pipe = build_pipeline(preprocessor, clf)
    
    # üöÄ Start Timer
    start_time = time.time() 
    
    try:
        # üöÄ Key Changes: using X_sample, y_sample and n_jobs=3
        res = cross_validate(pipe, X_sample, y_sample, cv=skf, scoring=scoring, n_jobs=3, return_train_score=False)
        
        # üöÄ End Timer and Calculate Duration
        end_time = time.time()
        duration = end_time - start_time
        
        results[name] = {
            'accuracy': res['test_accuracy'].mean(), 
            'f1': res['test_f1'].mean(), 
            'std': res['test_f1'].std(), 
            'time_sec': duration, # <-- Store the time
            'pipeline': pipe
        }
        
        print(f" -> acc {results[name]['accuracy']:.4f}, f1 {results[name]['f1']:.4f} (+/- {results[name]['std']:.4f})")
        # üü¢ Print the Time Result
        print(f" -> Time taken: {results[name]['time_sec']:.2f} seconds") 
        
    except Exception as e:
        print(f"    ERROR evaluating {name}:", e)
```

### Hyperparameter Tuning (LightGBM)

```python
# import time
from sklearn.model_selection import RandomizedSearchCV
from lightgbm import LGBMClassifier
from scipy.stats import randint as sp_randint, uniform as sp_uniform

# NOTE: This cell assumes you have run 'evaluation_setup.py' and have 
# X_sample, y_sample, preprocessor, skf, and scoring defined.

print("Starting Hyperparameter Tuning for LightGBM (LGBMClassifier)...")

# 1. Define the parameter grid for the search
lgbm_param_grid = {
    # Number of boosting stages (equivalent to number of trees)
    'classifier__n_estimators': sp_randint(100, 500), 
    # Step size shrinkage
    'classifier__learning_rate': sp_uniform(0.01, 0.15),
    # Maximum tree depth
    'classifier__max_depth': [-1, 6, 8, 10], # -1 means no limit
    # Max number of leaves in one tree. Crucial for model complexity.
    'classifier__num_leaves': sp_randint(20, 50),
    # Minimum number of data needed in a leaf
    'classifier__min_child_samples': sp_randint(15, 50),
    # Subsample ratio of the training instance
    'classifier__subsample': sp_uniform(0.6, 0.4), # range between 0.6 and 1.0
    # Subsample ratio of columns when constructing tree
    'classifier__colsample_bytree': sp_uniform(0.6, 0.4), # range between 0.6 and 1.0
}

# 2. Build the pipeline for tuning
lgbm_pipe = build_pipeline(preprocessor, LGBMClassifier(random_state=42, n_jobs=3, verbosity=-1))

# 3. Setup the Randomized Search
# n_iter=30 means we will test 30 random combinations of the hyperparameters
# This is a good balance between speed and finding an optimal set.
random_search = RandomizedSearchCV(
    lgbm_pipe,
    param_distributions=lgbm_param_grid,
    n_iter=30, # Number of parameter settings that are sampled
    scoring='f1', # Use F1 as the primary metric for optimization
    cv=skf,
    verbose=2,
    random_state=42,
    n_jobs=3, # Use 3 cores for parallel execution
    refit=True # Refit the best model on the whole X_sample
)

# 4. Run the tuning
start_time = time.time()
random_search.fit(X_sample, y_sample)
duration = time.time() - start_time

print("\n" + "="*50)
print(f"LightGBM Tuning Complete! Time taken: {duration:.2f} seconds")
print(f"Best F1 Score: {random_search.best_score_:.4f}")
print("Best Parameters Found:")
# Print parameters but remove the 'classifier__' prefix for cleaner output
best_params = {k.replace('classifier__', ''): v for k, v in random_search.best_params_.items()}
for k, v in best_params.items():
    print(f" - {k}: {v}")
print("="*50)

# Store the best estimator for later use
best_lgbm_estimator = random_search.best_estimator_
```
